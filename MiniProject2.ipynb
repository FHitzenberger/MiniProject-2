### Umgang mit Nullwerten

Fehlende Werte müssen sorgfältig behandelt werden, um die Datenintegrität zu gewährleisten. Eine gängige Methode ist das Auffüllen der fehlenden Werte mit dem Median des entsprechenden Attributs. Hier verwenden wir den Median, da er robust gegenüber Ausreißern ist und die Verteilung der Daten weniger stark beeinflusst als andere Methoden:

````python
df
````

````python
df.fillna(df.median(numeric_only=True), inplace=True)


print(df.head())
````

Nachdem die fehlenden Werte aufgefüllt wurden, überprüfen wir erneut, ob noch Nullwerte vorhanden sind:

````python
df.isnull().sum()
````

**Argumente für die Verwendung des Medians**
1. **Robustheit gegenüber Ausreißern:** Der Median ist weniger empfindlich gegenüber extremen Ausreißern als der Mittelwert. In Datensätzen mit Ausreißern kann der Mittelwert stark verzerrt sein, während der Median eine stabilere und repräsentativere zentrale Tendenz bietet.

2. **Erhalt der Verteilung:** Das Auffüllen mit dem Median verändert die Verteilung der Daten weniger stark als andere Methoden. Dies ist besonders wichtig, wenn die Verteilung nicht symmetrisch ist.

3. **Einfachheit und Effizienz:** Die Berechnung und Anwendung des Medians ist einfach und schnell, was in vielen Anwendungsfällen praktisch ist.

**Alternativen zum Auffüllen fehlender Werte**

1. **Auffüllen mit dem Mittelwert (Mean Imputation):**

- Pro: Einfach zu berechnen und anzuwenden.
- Contra: Kann durch Ausreißer verzerrt werden und die Varianz der Daten reduzieren.

2. **Auffüllen mit dem Modus (Mode Imputation):**

- Pro: Sinnvoll für kategoriale Daten.
- Contra: Nicht anwendbar für kontinuierliche Daten, und kann häufige Kategorien überrepräsentieren.

3. **Auffüllen durch Vorhersagemodelle (Predictive Imputation):**

- Pro: Nutzt die Beziehungen zwischen den Attributen, um genauere Schätzungen zu liefern.
- Contra: Komplexer und zeitaufwendiger, erfordert ein separates Modell.

4. **K-nächste Nachbarn Imputation (KNN Imputation):**

- Pro: Nutzt die Ähnlichkeit zwischen Beobachtungen, um fehlende Werte aufzufüllen.
- Contra: Rechenintensiv und erfordert eine sorgfältige Auswahl des Parameters k.

5. **Auffüllen mit speziellen Indikatoren:**

- Pro: Ermöglicht es, fehlende Werte explizit als solche zu markieren, was in einigen Modellen nützlich sein kann.
- Contra: Kann die Komplexität der Daten erhöhen und zu Interpretationsschwierigkeiten führen.

**Fazit**
Die Wahl der Methode zum Auffüllen fehlender Werte hängt stark vom spezifischen Datensatz und den Anforderungen der Analyse ab. In diesem Fall verwenden wir den Median, da er robust gegenüber Ausreißern ist und die Verteilung der Daten weniger stark verändert als andere Methoden. Dies hilft, die Integrität und die repräsentative Struktur der Daten aufrechtzuerhalten.

Nach der Datenbereinigung ist es nun an der Zeit, die Daten für die Modellierung vorzubereiten. Dies beinhaltet die Aufteilung der Daten in Trainings- und Testdaten sowie die Definition der Merkmale (Features) und der Zielvariable (Target).

## Aufbau von Pipelines für verschiedene Modelle

Im nächsten Schritt erstellen wir Pipelines für verschiedene Klassifikationsalgorithmen, um die Modelle effizient zu erstellen und zu validieren. Eine Pipeline fasst eine Sequenz von Datenverarbeitungsschritten und ein Modell zusammen.

Pipeline für SVM

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC


pipeline_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC())
])
````

Pipeline für Decision Tree

````python
from sklearn.tree import DecisionTreeClassifier

pipeline_dt = Pipeline([
    ('scaler', StandardScaler()),
    ('dt', DecisionTreeClassifier())
])
````

Pipeline für Decision Tree

````python
from sklearn.ensemble import RandomForestClassifier

pipeline_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier())
])
````

Pipeline für k-NN

````python
from sklearn.neighbors import KNeighborsClassifier


pipeline_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
````

Pipeline für Logistic Regression

````python
from sklearn.linear_model import LogisticRegression

pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression())
])
````

Pipelines kommen in ein Dictionary um einfachere Handhabung zu garantieren 

````python
pipelines = {
    'SVM': pipeline_svc,
    'Decision Tree': pipeline_dt,
    'Random Forest': pipeline_rf,
    'k-NN': pipeline_knn,
    'Logistic Regression': pipeline_lr
}
````

### Argumentation der Hyperparameter

#### SVM (Support Vector Machine)
- **`svc__C`: [0.1, 1, 10]**
  - **0.1**: Hohe Regularisierung, die das Modell bestraft, wenn es große Gewichte verwendet. Dies hilft, Überanpassung zu vermeiden.
  - **1**: Mittlere Regularisierung, die eine Balance zwischen Überanpassung und Unteranpassung bietet.
  - **10**: Geringe Regularisierung, die dem Modell mehr Freiheit gibt, komplexere Beziehungen in den Daten zu lernen.
- **`svc__gamma`: [1, 0.1, 0.01]**
  - **1**: Hohe Varianz, kleine Reichweite des Einflusses eines einzelnen Trainingsbeispiels.
  - **0.1**: Mittlere Varianz.
  - **0.01**: Niedrige Varianz, große Reichweite.

#### Decision Tree
- **`dt__max_depth`: [None, 10, 20, 30]**
  - **None**: Baum wächst, bis alle Blätter rein sind oder weniger als `min_samples_split` Knoten enthalten.
  - **10, 20, 30**: Begrenzung der Tiefe des Baumes, um Überanpassung zu vermeiden und die Komplexität des Modells zu kontrollieren.

#### Random Forest
- **`rf__n_estimators`: [50, 100, 200]**
  - **50**: Geringere Anzahl an Bäumen, schnelleres Training, aber möglicherweise geringere Stabilität.
  - **100**: Standardanzahl, bietet gute Balance zwischen Leistung und Trainingszeit.
  - **200**: Höhere Anzahl an Bäumen, kann zu besserer Leistung führen, erfordert aber mehr Rechenressourcen.

#### k-NN (k-Nearest Neighbors)
- **`knn__n_neighbors`: [3, 5, 7]**
  - **3**: Weniger Nachbarn, mehr Sensibilität gegenüber Rauschen.
  - **5**: Typischer Wert, bietet gute Balance zwischen Stabilität und Sensibilität.
  - **7**: Mehr Nachbarn, stabilere Entscheidungen, aber kann glattere Entscheidungsgrenzen zur Folge haben.

#### Logistic Regression
- **`lr__C`: [0.1, 1, 10]**
  - **0.1**: Stärkere Regularisierung, verhindert Überanpassung.
  - **1**: Mittlere Regularisierung.
  - **10**: Geringere Regularisierung, lässt das Modell komplexere Beziehungen lernen.
  
  
Für jede Pipeline definieren wir ein Parameterraster für die Hyperparameteroptimierung. Die Argumentation der Hyperparameter wird ebenfalls erläutert:

````python
param_grids = {
    'SVM': {'svc__C': [0.1, 1, 10], 'svc__gamma': [1, 0.1, 0.01]},
    'Decision Tree': {'dt__max_depth': [None, 10, 20, 30]},
    'Random Forest': {'rf__n_estimators': [50, 100, 200]},
    'k-NN': {'knn__n_neighbors': [3, 5, 7]},
    'Logistic Regression': {'lr__C': [0.1, 1, 10]}
}
````

# 6 Vergleich der Klassifizierungsalgorithmen

Wir führen GridSearchCV für jede Pipeline aus, um die besten Hyperparameter zu finden:

````python
import time


best_estimators = {}
for algo, pipeline in pipelines.items():
    print(f"Optimizing {algo}...")
    start = time.time()
    grid = GridSearchCV(pipeline, param_grids[algo], cv=5)
    grid.fit(X_train, y_train)
    best_estimators[algo] = grid.best_estimator_
    end = time.time()
    print(f"{algo} took {end - start:.2f} seconds")
````

Die Ergebnisse zeigen die Zeit, die benötigt wurde, um für jeden Klassifikationsalgorithmus die Hyperparameter-Optimierung mittels `GridSearchCV` mit 5-facher Kreuzvalidierung durchzuführen. Hier ist die detaillierte Interpretation und Beschreibung der Ergebnisse:

### SVM (Support Vector Machine)
- **Zeit:** 16.74 Sekunden
- **Interpretation:**

Wir erstellen einen DataFrame zur Evaluierung der Modelle basierend auf verschiedenen Metriken: Genauigkeit (Accuracy), F1-Score, Präzision (Precision), Recall und den Matthews-Korrelation-Koeffizienten (MCC):

````python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, matthews_corrcoef

results = {}
for algo, model in best_estimators.items():
    y_pred = model.predict(X_test)
    results[algo] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'MCC': matthews_corrcoef(y_test, y_pred)
    }
````

````python
results_df = pd.DataFrame(results).T
print("Evaluation Metrics for Optimized Models:")
print(results_df)

# Best model
best_model_name = results_df['Accuracy'].idxmax()
print(f"The best model is: {best_model_name} with an accuracy of {results_df.loc[best_model_name, 'Accuracy']:.2f}")
````

### Interpretation der Metriken:

1. **Accuracy (Genauigkeit)**
   - **Bedeutung:** Der Anteil der korrekt vorhergesagten Beobachtungen an den Gesamtbeobachtungen.
   - **Ergebnisse:** Random Forest hat die höchste Genauigkeit (0.994872), was bedeutet, dass fast alle Vorhersagen korrekt waren. k-NN hat die niedrigste Genauigkeit (0.916667), was darauf hinweist, dass dieses Modell mehr Fehler gemacht hat als die anderen.

2. **F1 Score**
   - **Bedeutung:** Der harmonische Mittelwert von Präzision und Recall. Ein hoher F1-Score zeigt an, dass das Modell sowohl eine hohe Präzision als auch einen hohen Recall hat.
   - **Ergebnisse:** Random Forest hat auch hier den höchsten Wert (0.994792), gefolgt von Decision Tree (0.987080). k-NN zeigt den niedrigsten F1-Score (0.909091), was darauf hinweist, dass die Balance zwischen Präzision und Recall schlechter ist als bei den anderen Modellen.

3. **Precision (Präzision)**
   - **Bedeutung:** Der Anteil der richtig vorhergesagten positiven Beobachtungen an allen vorhergesagten positiven Beobachtungen.
   - **Ergebnisse:** SVM, Random Forest und Logistic Regression haben eine Präzision von 1.000000, was bedeutet, dass alle positiven Vorhersagen korrekt waren. k-NN hat eine Präzision von 0.987842, was dennoch sehr hoch ist.

4. **Recall**
   - **Bedeutung:** Der Anteil der richtig vorhergesagten positiven Beobachtungen an allen tatsächlichen positiven Beobachtungen.
   - **Ergebnisse:** Decision Tree und Random Forest zeigen den höchsten Recall (0.989637), was darauf hinweist, dass diese Modelle die meisten tatsächlichen positiven Fälle korrekt erkannt haben. k-NN hat den niedrigsten Recall (0.841969), was bedeutet, dass es mehr tatsächliche positive Fälle verpasst hat.

5. **MCC (Matthews-Korrelation-Koeffizient)**
   - **Bedeutung:** Eine Metrik, die die Qualität der binären (zwei Klassen) Klassifikationen misst. Sie berücksichtigt alle vier Elemente der Verwirrungsmatrix (True Positives, True Negatives, False Positives, False Negatives) und liefert einen Wert zwischen -1 und +1, wobei +1 eine perfekte Vorhersage, 0 keine bessere Vorhersage als Zufall und -1 eine totale Fehleinschätzung darstellt.
   - **Ergebnisse:** Random Forest zeigt den höchsten MCC (0.989794), gefolgt von Decision Tree (0.974370). Dies deutet darauf hin, dass diese Modelle eine sehr gute Gesamtleistung haben. k-NN hat den niedrigsten MCC (0.842138), was zeigt, dass die Vorhersagequalität im Vergleich zu den anderen Modellen schlechter ist.

### Zusammenfassung:

- **Random Forest**: Hervorragende Leistung in allen Metriken, besonders in Genauigkeit, F1-Score und MCC. Dies ist ein Zeichen dafür, dass dieses Modell sehr gut für die gegebene Aufgabe geeignet ist.
- **Decision Tree**: Zeigt ebenfalls eine sehr gute Leistung, nur leicht hinter dem Random Forest. Dies kann darauf hinweisen, dass die Entscheidungsbaum-Struktur gut zu den Daten passt.
- **SVM und Logistic Regression**: Beide Modelle zeigen ebenfalls sehr gute Ergebnisse, besonders in der Präzision. Dies bedeutet, dass sie sehr gut darin sind, positive Fälle korrekt zu identifizieren.
- **k-NN**: Hat in allen Metriken die niedrigsten Werte, was darauf hinweist, dass dieses Modell weniger geeignet ist für diese spezifische Aufgabe oder dass die Hyperparameter weiter optimiert werden könnten.
